```python
cmap = sns.cubehelix_palette(start=5, light=1, as_cmap=True)

yticks = np.linspace(0,1,6)

x_end = 100
xticks = np.linspace(0,100,5)

ax = sns.heatmap(data, linewidth=0, xticklabels=xticks,
                 yticklabels=yticks[::-1], square=True, cmap=cmap)
ax.set_xticks(xticks*ax.get_xlim()[1]/(x_end))
ax.set_yticks(yticks*ax.get_ylim()[1])


fig = ax.get_figure()
fig.savefig("output.pdf")
```
